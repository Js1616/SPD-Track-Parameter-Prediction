```python
import numpy as np 
import pandas as pd 
from sklearn.preprocessing import StandardScaler
import joblib
```

```python
load_data = r"C:\Projects\SPD-Track-Parameter-Prediction\data\processed\processed_signal_hits.tsv"
```

```python
df = pd.read_csv(load_data, sep='\t') 
```

```python
print(df.dtypes)
```

```python
duplicates = df.duplicated().sum()
```

```python
df = df.sort_values(
    by=["event_id", "track_id", "station"]
)

df = df.reset_index(drop=True)
```

```python
# Additional physics based features

df["phi_hit"] = np.arctan2(
    df["y"],
    df["x"]
)


df["rho_z"] = np.sqrt(
    df["radius"]**2 +
    df["z"]**2
)
```

```python
feature_columns = [
"x",
"y",
"z",

]

target_columns = ["px","py","pz"]
```

```python
feature_scaler = StandardScaler()

df[feature_columns] = feature_scaler.fit_transform(
    df[feature_columns]
)
station_column = ["station"]
target_scaler = StandardScaler()

df[target_columns] = target_scaler.fit_transform(
    df[target_columns]
)
```

```python
print(df[feature_columns].describe())
```

```python
print(df[target_columns].describe())
```

```python
joblib.dump(
    feature_scaler,
    r"c:\Projects\SPD-Track-Parameter-Prediction\notebooks\feature_scaler.pkl"
)

joblib.dump(
    target_scaler,
    r"c:\Projects\SPD-Track-Parameter-Prediction\notebooks\target_scaler.pkl"
)
```

```python
df.to_csv(
    r"c:\Projects\SPD-Track-Parameter-Prediction\notebooks\processed_signal_hits_scaled.tsv",
    sep="\t",
    index=False
)
```
